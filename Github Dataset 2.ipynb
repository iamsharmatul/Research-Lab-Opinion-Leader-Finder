```python
def get_user(df):
    users = pd.DataFrame()
    for x in range(len(df['items'])):
        y = list(df['items'][x])
        users = users.append(y, ignore_index=True)
    return users
```

```python
users = get_user(df)
```

```python
def fetch_git_data(url):
    pages = []
    url = url + "?per_page=100"
    res=requests.get(url,headers={"Authorization": "0cb6df3524a429846abcfac4db1ea81e12ac2f4f"})
    data=res.json()
    pages.extend(data)
    counter = 0
    #print (res.links.keys())
    while 'next' in res.links.keys():
        counter += 1
        res = requests.get(res.links['next']['url'], headers={"Authorization": "0cb6df3524a429846abcfac4db1ea81e12ac2f4f"})
        pages.extend(res.json())
    #print(len(pages))
    return pages
```

```python
data = list(map(fetch_git_data ,list(users.followers_url)))
print ('done')
```
